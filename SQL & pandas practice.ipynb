# Group by

## Average salary

```python
sql = """
SELECT DEPTNO,AVG(SAL)
FROM EMP
"""
pd.read_sql_query(sql, mydb)
```

```python
emp_df.SAL.mean()
```

## Average salary by department number

```python
sql = """
SELECT DEPTNO,AVG(SAL)
FROM EMP
GROUP BY DEPTNO;
"""
pd.read_sql_query(sql, mydb)
```

```python
emp_df[["DEPTNO", "SAL"]].groupby(["DEPTNO"]).mean()
```

## Total number of employees and the number of employees receiving commissions by department number

```python
sql = """
SELECT DEPTNO,COUNT(*),COUNT(COMM)
FROM EMP
GROUP BY DEPTNO;
"""
pd.read_sql_query(sql, mydb)
```

```python
emp_df[["DEPTNO", "EMPNO", "COMM"]].groupby(["DEPTNO"]).count()
```

## Maximum salary and minimum salary by department number

```python
sql = """
SELECT DEPTNO,MAX(SAL),MIN(SAL)
FROM EMP
GROUP BY DEPTNO;
"""
pd.read_sql_query(sql, mydb)
```

```python
emp_df[["DEPTNO", "SAL"]].groupby(["DEPTNO"]).max()
```

```python
emp_df[["DEPTNO", "SAL"]].groupby(["DEPTNO"]).min()
```

```python
emp_df[["DEPTNO", "SAL"]].groupby(["DEPTNO"]).describe()
```

## Salary averages by department number, but only those with a result of 2000 or higher

```python
sql = """
SELECT DEPTNO,AVG(SAL)
FROM EMP
GROUP BY DEPTNO
HAVING AVG(SAL) >= 2000;
"""
pd.read_sql_query(sql, mydb)
```

```python
emp_df[["DEPTNO", "SAL"]].groupby(["DEPTNO"]).mean()
```

```python
temp = emp_df[["DEPTNO", "SAL"]].groupby(["DEPTNO"]).mean()
temp[temp.SAL >= 2000]
```

## Salary averages by department number only for employees whose salary is 1000 or higher (only records with an average of 2000 or more)

```python
sql = """
SELECT DEPTNO,AVG(SAL)
FROM EMP
WHERE SAL >= 1000
GROUP BY DEPTNO
HAVING AVG(SAL) >= 2000;
"""
pd.read_sql_query(sql, mydb)
```

```python
temp = emp_df[emp_df.SAL >= 1000][["DEPTNO", "SAL"]].groupby(["DEPTNO"]).mean()
temp[temp.SAL >= 2000]
```

## Employee name, salary, and salary grade

```python
sql = """
SELECT E.ENAME,E.SAL,S.GRADE
FROM EMP E,SALGRADE S
WHERE E.SAL >= S.LOSAL AND E.SAL <= S.HISAL;
"""
pd.read_sql_query(sql, mydb)
```

```python
# the same as above
sql = """
SELECT E.ENAME,E.SAL,S.GRADE
FROM EMP E,SALGRADE S
WHERE E.SAL BETWEEN S.LOSAL AND S.HISAL;
"""
pd.read_sql_query(sql, mydb)
```

```python
def temp_grd(row):
    if row["SAL"] > 700 and row["SAL"] <= 1200:
        grd = 1
    elif row["SAL"] > 1201 and row["SAL"] <= 1400:
        grd = 2
    elif row["SAL"] > 1401 and row["SAL"] <= 2000:
        grd = 3
    elif row["SAL"] > 2001 and row["SAL"] <= 3000:
        grd = 4
    elif row["SAL"] > 3001 and row["SAL"] <= 9999:
        grd = 5
    return grd

emp_df["GRADE"] = emp_df.apply(temp_grd, axis=1)
emp_df[["ENAME", "SAL", "GRADE"]]
```

```python
emp_df = emp_df.drop("GRADE", axis=1)
```

```python
# the same as above
sql = """
SELECT C.ENAME,C.DEPTNO
FROM EMP E INNER JOIN EMP C ON E.DEPTNO = C.DEPTNO 
WHERE E.EMPNO <> C.EMPNO
AND E.ENAME = 'ALLEN'
ORDER BY C.ENAME;
"""
pd.read_sql_query(sql, mydb)
```

## Employee name and department name of an employee in a department with department number 10

```python
sql = """
SELECT E.ENAME, D.DNAME
FROM EMP E, DEPT D
WHERE E.DEPTNO = D.DEPTNO
AND D.DEPTNO = 10;
"""
pd.read_sql_query(sql, mydb)
```

```python
# the same as above
sql = """
SELECT E.ENAME, D.DNAME
FROM EMP E,
(
    SELECT DEPTNO,DNAME
    FROM DEPT
    WHERE DEPTNO = 10
) D
WHERE E.DEPTNO = D.DEPTNO;
"""
pd.read_sql_query(sql, mydb)
```

```python
pd.merge(emp_df, dept_df[dept_df.DEPTNO ==10])[["ENAME","DNAME"]]
```

## Employee numbers, names, and salaries of employees who receive more than average salary

```python
sql = """
SELECT EMPNO,ENAME,SAL
FROM EMP
WHERE SAL > (SELECT AVG(SAL)
          FROM EMP)
ORDER BY SAL DESC;
"""
pd.read_sql_query(sql, mydb)
```

```python
emp_df[emp_df.SAL > emp_df.SAL.mean()].sort_values(by="SAL", ascending=[False])[["EMPNO", "ENAME", "SAL"]]
```

## Employee number and employee name of the employee with the highest salary among the employees with the department number 10

```python
sql = """
SELECT EMPNO,ENAME
FROM EMP
WHERE SAL = (SELECT MAX(SAL) 
         FROM EMP 
         WHERE DEPTNO = 10);
"""
pd.read_sql_query(sql, mydb)
```

```python
emp_df[emp_df.DEPTNO == 10][emp_df[emp_df.DEPTNO == 10].SAL == emp_df[emp_df.DEPTNO == 10].SAL.max()][["EMPNO", "ENAME"]]
```
